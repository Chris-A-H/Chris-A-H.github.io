```python
import numpy as np
import matplotlib.pyplot as plt

days = np.arange(1, 31)
solar_output = np.random.uniform(5, 15, size=30)  # Simulated solar production
wind_output = np.random.uniform(2, 8, size=30)    # Simulated wind production
home_usage = np.random.uniform(10, 12, size=30)   # Simulated home demand

plt.plot(days, solar_output, label='Solar Output (kWh)')
plt.plot(days, wind_output, label='Wind Output (kWh)')
plt.plot(days, home_usage, label='Home Consumption (kWh)', linestyle='--')
plt.xlabel('Day')
plt.ylabel('Energy (kWh)')
plt.title('Energy Generation vs Consumption')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 30 days of production again
days = np.arange(1, 31)
solar_output = np.random.uniform(5, 15, size=30)  # kWh/day
wind_output = np.random.uniform(2, 8, size=30)    # kWh/day

# Total energy produced from each source
total_solar = np.sum(solar_output)
total_wind = np.sum(wind_output)

# Pie chart data
labels = ['Solar Energy', 'Wind Energy']
sizes = [total_solar, total_wind]
explode = (0.05, 0)  # Slightly "explode" solar slice for emphasis

# Plot
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
ax.axis('equal')  # Equal aspect ratio ensures pie chart is a circle

plt.title('Energy Source Contribution Over 30 Days')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 30 days of weather
days = np.arange(1, 31)
weather = np.random.choice(['Sunny', 'Cloudy'], size=30, p=[0.7, 0.3])  # 70% Sunny, 30% Cloudy

# Simulated production based on weather
solar_output = np.array([np.random.uniform(12, 15) if w == 'Sunny' else np.random.uniform(4, 7) for w in weather])
wind_output = np.random.uniform(2, 8, size=30)  # Wind output remains variable

# Total energy produced
total_energy_generated = np.sum(solar_output + wind_output)

# Energy usage assumptions (adjusting for weather):
# On sunny days, more energy is used directly because production matches daytime demand better
# On cloudy days, more needs to be stored or bought
energy_used_directly = np.sum([0.7 * (s+w) if weather[i]=='Sunny' else 0.4 * (s+w) for i, (s, w) in enumerate(zip(solar_output, wind_output))])
energy_stored_in_battery = np.sum([0.25 * (s+w) if weather[i]=='Sunny' else 0.4 * (s+w) for i, (s, w) in enumerate(zip(solar_output, wind_output))])
energy_sold_to_market = total_energy_generated - (energy_used_directly + energy_stored_in_battery)

# Prepare Pie Chart Data
labels = ['Used by Home', 'Stored in Battery', 'Sold to Market']
sizes = [energy_used_directly, energy_stored_in_battery, energy_sold_to_market]
colors = ['#4CAF50', '#2196F3', '#FFC107']
explode = (0.05, 0.05, 0.05)

# Plot
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
plt.title('Dynamic Breakdown of Energy Usage Over 30 Days (Sunny vs Cloudy)')
plt.show()

# (Optional) Show how many sunny vs cloudy days
print(f"Sunny Days: {np.sum(weather=='Sunny')}, Cloudy Days: {np.sum(weather=='Cloudy')}")
```
